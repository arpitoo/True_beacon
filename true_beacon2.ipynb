```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('FINALE.csv', header=0, skipinitialspace=True)
```

```python
data.columns = ['date', 'banknifty', 'nifty', 'bankclose', 'niftyclose']
```

```python
data
```

```python
data = data[['bankclose','banknifty',	'nifty','niftyclose']].interpolate(method='linear')
```

```python
data['spread'] = data['banknifty'] - data['nifty']
```

```python
data['ratio'] = data['bankclose'] / data['niftyclose']
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame with columns 'spread' and 'ratio'
# 'spread' is the independent variable and 'ratio' is the dependent variable

# Extracting independent variable (X) and dependent variable (y)
X = data[['spread']]  # Independent variable (spread)
y = data['ratio']     # Dependent variable (ratio)

# Create a Linear Regression model
model = LinearRegression()

# Fit the model using your data
model.fit(X, y)

# Print the coefficients
print("Intercept:", model.intercept_)
print("Coefficient (slope):", model.coef_[0])
```

```python
predicted_values = model.predict(X)

# Add the predicted values as a new column in your DataFrame
data['predicted_ratio'] = predicted_values
```

```python
data
```

```python
# Setting the window size to 375 minutes (1 trading day)
window_size = 30

# Calculate the rolling mean and standard deviation for the spread
data['ratio_Mean'] = data['ratio'].rolling(window=window_size).mean()
data['ratio_Std'] = data['ratio'].rolling(window=window_size).std()
```

```python
data['error'] = data['predicted_ratio'] - data['ratio']
```

```python
data
```

```python
data['tte'] = 27
```

```python
data['theoritical_p/l'] = data['spread'] * (data['tte']**0.7)
data['p/l'] = data['theoritical_p/l']
```

```python
import pandas as pd
import math

# Assuming 'data' is your DataFrame
profit = 0

# Iterate over rows of the DataFrame
for index, row in data.iterrows():
    value = row['error']
    ratio_Std = row['ratio_Std']

    # Check if 'error' value or 'ratio_Std' value is NaN
    if math.isnan(value) or math.isnan(ratio_Std):
        continue
    elif value > ratio_Std:
        profit += row['p/l']
    elif -value > ratio_Std:
        profit -= row['p/l']

print(profit)
```
